```python
import pandas as pd
```

```python
df=pd.read_csv('titanic_for_dtree.csv')
df.head()
```

```python
target =df['Survived']
```

```python
input=df[['Pclass','Sex','Age',"Fare"]]
input
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python

le_Sex = LabelEncoder()
```

```python
input['Sex_n'] = le_Sex.fit_transform(input['Sex'])
input
```

```python
input = input.drop(['Sex'],axis='columns')
```

```python
input
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input,target)
```

```python
model.score(input,target)
```

```python
model.predict([[1,38.0,71.2833,0]])
```
