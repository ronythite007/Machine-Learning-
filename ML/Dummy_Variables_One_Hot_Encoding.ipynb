```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('dummyhomeprices.csv')
df
```

```python
pd.get_dummies(df.town) 
```

```python
dummies = pd.get_dummies(df['town'])
dummies = dummies.astype(int)  #changing into integer
dummies
```

```python
merge = pd.concat([df,dummies],axis='columns')
merge
```

```python
final = merge.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x=final.drop(['price'],axis='columns')
x
```

```python
y= final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2500,0,1]])
```

```python
accuracy = model.score(x,y)* 100 # gives the accuracy of model
accuracy
```
