### label encoding

```python
df4 = df3.copy() # copy all data from the df3
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4)
df5 =df5.astype(int)
df5
```

```python
x = df5.drop('HeartDisease',axis='columns')
y = df5.HeartDisease
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(x_train, y_train)
model_rf.score(x_test, y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
```

```python
x_pca = pca.fit_transform(x)
```

```python
x_pca.shape
```

```python
x_train_pca,x_test_pca,y_train_pca,y_test_pca = train_test_split(x_pca,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier()
```

```python
model_rf.fit(x_train_pca,y_train_pca)
```

```python
model_rf.score(x_test_pca,y_test_pca)
```
