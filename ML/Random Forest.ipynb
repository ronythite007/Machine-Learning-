# Random Forest

Random forest is a commonly-used machine learning algorithm, trademarked by Leo Breiman and Adele Cutler, that combines the output of multiple decision trees to reach a single result. 

### Digits DataSet

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits 
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
df = pd.DataFrame(digits.data)
```

```python
df
```

```python
df["Target"]=digits.target
```

```python
df
```

```python
x=df.drop(["Target"],axis="columns")
```

```python
x
```

```python
y = digits.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
digits.data[100]
```

```python
digits.target[100]
```

```python
model.predict([digits.data[56]])
```

```python
accuracy = model.score(x_test,y_test)
```

```python
import math
print(f"Accuracy of Random Forest : {round(accuracy * 100,2)} ")
```
